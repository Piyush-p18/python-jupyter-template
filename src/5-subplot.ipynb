### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Creating the subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, color='blue', marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y = X')
axs[0, 0].grid(True)

# Bar plot
axs[0, 1].bar(categories, values, color=['skyblue', 'lightgreen', 'lightcoral', 'gold'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', label='Data Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Display
plt.show
```
